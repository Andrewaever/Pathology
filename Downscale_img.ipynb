```python
from patchify import patchify
import matplotlib.pylab as plt
import cv2
import numpy as np
import glob
from PIL import Image
```

```python
filenames = glob.glob("/Users/frederiklindvig/Desktop/PathologyInput/*.jpeg")
#filenames.sort()
images = [cv2.imread(img) for img in filenames]

image_list = []

for img in images:
  image_list.append(img)
```

```python
scale_down = 0.1
scaled_imgs =[]

for img in image_list:
  scaled_f_down = cv2.resize(img, None, fx= scale_down, fy= scale_down, interpolation= cv2.INTER_LINEAR)
  scaled_imgs.append(scaled_f_down)
```

```python
plt.figure(figsize=(10,10))

for i in range(90):
  plt.subplot(10,10,i+1)
  plt.imshow(scaled_imgs[i])
  plt.axis('off')
```
